```python
import numpy as np 
import pandas as pd 
```

```python
movies = pd.read_csv('tmdb_5000_credits.csv')
credits = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def make(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L  
```

```python
movies['genres']= movies['genres'].apply(make)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(make)
```

```python
def make3(obj):
    L = []
    counter = 0 
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1 
        else:
            break
    return L  
```

```python
movies['cast']= movies['cast'].apply(make3)
```

```python
 def get_director(obj):
        L= []
        for i in ast.literal_eval(obj):
            if i['job'] == 'Director':
                L.append(i['name'])
                break
        return L
```

```python
movies['crew']= movies['crew'].apply(get_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] 
+ movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tag']]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors
```

```python
feature_names = cv.get_feature_names_out()
for name in feature_names:
    print(name)
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    try:
        # Find the index of the movie in the DataFrame
        movie_index = new_df[new_df['title'] == movie].index[0]

        # Get the similarity scores for the movie
        distances = similarity[movie_index]

        # Create a list of tuples with movie indices and their similarity scores
        movie_list = list(enumerate(distances))

        # Sort the list based on similarity scores in descending order
        movie_list = sorted(movie_list, key=lambda x: x[1], reverse=True)

        # Get the top 5 similar movies (excluding the first one which is the input movie itself)
        recommended_movies = movie_list[1:6]

        # Extract and print the movie titles of the recommended movies
        for i in recommended_movies:
            print(new_df.iloc[i[0]].title)
    except IndexError:
        print("Movie not found in the dataset.")
    except Exception as e:
        print(f"An error occurred: {e}")


```

```python
recommend('Iron Man')
```
